```python
import can
import udsoncan
```

```python
bus = can.Bus(interface='socketcan',
              channel='can0',
              receive_own_messages=False)

# send a message
message = can.Message(arbitration_id=0x110, is_extended_id=False,
                      data=[0x8f, 0x00, 0x00, 0x9e, 0x86, 0x00, 0x00, 0x00])
bus.send(message, timeout=0.2)
```

```python
for X in range(0x00, 0xFF):
    print(f"{X:X}")
    message = can.Message(arbitration_id=0x110, is_extended_id=False,
                      data=[0X00, X, 0x00, 0x9e, 0x86, 0x00, 0x00, 0x00])
    bus.send(message, timeout=0.2)
    time.sleep(0.25)
```
